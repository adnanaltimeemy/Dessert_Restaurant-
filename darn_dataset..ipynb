```python
import pandas as pd
from datetime import datetime

# Define the schema columns
columns = [
    "region",
    "period",
    "total_sales_usd",
    "cost_of_goods_sold_usd",
    "gross_profit_usd",
    "gross_margin_pct",
    "operating_expenses_usd",
    "operating_profit_usd",
    "operating_margin_pct",
    "units_sold",
    "average_selling_price_usd",
    "customer_count",
    "repeat_customer_rate_pct",
    "median_customer_age",
    "gender_share_male_pct",
    "gender_share_female_pct",
    "top_3_cities_by_sales",
    "notes"
]

# Create empty dataframes for UK and US
df_uk = pd.DataFrame(columns=columns)
df_us = pd.DataFrame(columns=columns)

# Example helper to calculate derived fields
def add_row(df, region, period, total_sales, cogs, op_expenses,
            units_sold=None, customer_count=None, repeat_rate=None,
            median_age=None, male_pct=None, female_pct=None,
            top_cities=None, notes=None):
    gross_profit = total_sales - cogs
    gross_margin = (gross_profit / total_sales * 100) if total_sales else None
    operating_profit = gross_profit - op_expenses if op_expenses is not None else None
    operating_margin = (operating_profit / total_sales * 100) if total_sales and operating_profit is not None else None
    avg_selling_price = (total_sales / units_sold) if units_sold else None
    
    df.loc[len(df)] = [
        region,
        period,
        total_sales,
        cogs,
        gross_profit,
        gross_margin,
        op_expenses,
        operating_profit,
        operating_margin,
        units_sold,
        avg_selling_price,
        customer_count,
        repeat_rate,
        median_age,
        male_pct,
        female_pct,
        top_cities,
        notes
    ]

# Add a couple of sample rows for illustration
add_row(
    df_uk, "UK", "2024-12", total_sales=500000, cogs=200000,
    op_expenses=150000, units_sold=20000, customer_count=8000,
    repeat_rate=25.0, median_age=35, male_pct=55.0, female_pct=45.0,
    top_cities="London|Manchester|Bristol",
    notes="Placeholder data; replace with real figures"
)

add_row(
    df_us, "US", "2024-12", total_sales=2000000, cogs=800000,
    op_expenses=600000, units_sold=80000, customer_count=25000,
    repeat_rate=30.0, median_age=38, male_pct=52.0, female_pct=48.0,
    top_cities="Burlington|Denver|Portland",
    notes="Placeholder data; replace with real figures"
)

# Export to CSV
df_uk.to_csv("darn_tough_uk_sales_demo.csv", index=False)
df_us.to_csv("darn_tough_us_sales_demo.csv", index=False)

print("CSV files generated: darn_tough_uk_sales_demo.csv, darn_tough_us_sales_demo.csv")
```
